```python
### Train OTB vector classifier ###
### Source: https://www.orfeo-toolbox.org/CookBook/Applications/app_TrainVectorClassifier.html ###

import sys
# Adapt paths
# Path to the OTB Python folder
sys.path.append('')
import os
otbpath = os.environ.get('OTB_APPLICATION_PATH', None)
# Path to OTB applications
if otbpath is None:
    os.environ['OTB_APPLICATION_PATH']=''

import otbApplication

# Set general input parameters
# Input shapefile to be classified
in_features = ''
# Training shapefile
train_features = ''
# Name for the model
modelname = ''
# Filename for the confusion matrix (if no separate validation data are given, the model is tested using the training areas!)
confmatname = ''
# String containing space separated column names of attributes that should be taken into account. Order matters.
attributestring = ''
# Filename for the classification output
outname = ''

# The following line creates an instance of the TrainVectorClassifier application
TrainVectorClassifier = otbApplication.Registry.CreateApplication('TrainVectorClassifier')

# The following lines set all the application parameters:
TrainVectorClassifier.SetParameterStringList('io.vd', [train_features])
#TrainVectorClassifier.SetParameterString('io.stats', 'meanVar.xml')
TrainVectorClassifier.SetParameterString('cfield', 'id')
TrainVectorClassifier.SetParameterString('io.out', modelname)
#Choose classification model here
TrainVectorClassifier.SetParameterString('classifier', 'ann')
TrainVectorClassifier.SetParameterString('classifier.ann.sizes', '5')
#List of attributes to use for classification. Order matters.
TrainVectorClassifier.SetParameterString('feat', attributestring)
TrainVectorClassifier.SetParameterString('io.confmatout', confmatname)

# The following line execute the application
TrainVectorClassifier.ExecuteAndWriteOutput()


# The following line creates an instance of the VectorClassifier application
VectorClassifier = otbApplication.Registry.CreateApplication('VectorClassifier')

# The following lines set all the application parameters:
VectorClassifier.SetParameterString('in', in_features)
#VectorClassifier.SetParameterString('instat', 'meanVar.xml')
VectorClassifier.SetParameterString('model', modelname)
VectorClassifier.SetParameterString('feat', attributestring)
VectorClassifier.SetParameterString('cfield', 'predicted')
VectorClassifier.SetParameterString('out', outname)

# The following line execute the application
VectorClassifier.ExecuteAndWriteOutput()
```
